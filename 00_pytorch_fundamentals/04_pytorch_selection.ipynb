## Indexing (Selecting data from tensor)

```python
# creating tensor

x = torch.arange(1,10).reshape(1,3,3)
x, x.shape
```

```python
# selecting the first index

x[0]
```

```python
x[0][1]
```

```python
x[0][1][1]
```

```python
# Use ':' to select all of the targeted demensions
x[:,2]
```

```python
# Get all values of 0th and 1st dimentions but only index 1 and 2nd dimentions

x[:,:,1]
```

```python
# Get all values of the 0th dimentions but only of the 1 index value of 1st and 2nd dimentions

x[:,1,1]
```

```python
# Index on x to return 9
x[0,2,2]

# Index of x to return 3,6,9
x[0,:,2]
```

## Pytorch tensor to Numpy

```python
# Numpy array to tensor
import torch
import numpy as np

array = np.arange(1,10)
tensor = torch.from_numpy(array)

array, tensor
```

```python
# Tensor to Numpy array
tensor = torch.ones(10)
numpy_tensor = tensor.numpy()

tensor, numpy_tensor
```
