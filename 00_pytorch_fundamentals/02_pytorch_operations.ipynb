```python
import torch
```

### Tensor Manipulation

Tensor operations include:
* Addition
* Subtraction
* Multiplication(element-wise)
* Division
* Matrix Multiplication

```python
# Tensor creation

tensor = torch.tensor([1,2,3])

tensor
```

```python
# Tensor Addition

torch.add(tensor, 10)

tensor + 10
```

```python
# Tensor Multiply

torch.mul(tensor, 10)

tensor * 10
```

```python
# Tensor Subtract

torch.sub(tensor, 10)

tensor - 10
```

### Matrix Multiplication

```python
# Element wise multiplication 

print(tensor,'*', tensor)
print(f"Equals: {tensor * tensor}")
```

```python
# Matrix Multiplication

torch.matmul(tensor, tensor)
```

```python
tensor_A = torch.rand(2,3)
tensor_B = torch.rand(2,3)
```

```python
torch.matmul(tensor_A, tensor_B)
```

```python
print(tensor_A.shape)
print(tensor_A.T.shape)
```

```python
# The matrix multiplication works when matrix b in Transposed

torch.matmul(tensor_A, tensor_B.T)
```

### Aggregation functions

```python
# Creating a tensor

x = torch.arange(0,100,10)
x
```

```python
# Find the min

x.min(), torch.min(x)
```

```python
# Find the max

x.max(), torch.max(x)
```

```python
# Find the mean

x.type(torch.float32).mean(), torch.mean(x.type(torch.float32))
```

```python
# Find the sum
x.sum(), torch.sum(x)
```

### Finding the positional min max

```python
x
```
